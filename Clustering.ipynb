# Clustering
Obtención de clusters para segmentar usuarios con consumos parecidos.

## Ejemplo 3D
### Entrenamiento
Es necesario obtener el valor de K (número de clusters a usar). Este valor se obtine de manera empírica probando con distintos valores e identificacndo el valor desde el cual ya no se producen mejoras sustanciales.

#### Configuración y carga del dataset

```python
import pandas as pd
import numpy as np

#Seccion de parametrizacion
#--------------------------

__fichero = "./datasets/datos_clustering_3D.csv"
__K_min = 1
__K_max = 50
__K_salto=1

#***************************

dataframe = pd.read_csv(__fichero)
print (dataframe.head(5))
dataframe.fillna(0, inplace=True)
X = np.array(dataframe[["H0-6","H7-14","H15-23"]])
```

#### Visualización de los datos cargados

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)

ax.scatter(X[:, 0], X[:, 1], X[:, 2],s=60)
```

#### Estimación del valor de K

```python
from sklearn.cluster import KMeans
fig = plt.figure()
Nc = range(__K_min, __K_max, __K_salto)
kmeans = [KMeans(n_clusters=i) for i in Nc]
score = [kmeans[i].fit(X).inertia_ for i in range(len(kmeans))]
plt.rcParams['figure.figsize'] = (16, 9)
plt.plot(Nc,score)
plt.xlabel('K')
plt.ylabel('Score')
plt.show()
```

### Ejecución

#### Ejecución del Knn

```python
__K = 6

kmeans = KMeans(n_clusters=__K).fit(X)
centroids = kmeans.cluster_centers_
print(centroids)
```

#### Asiganmos cada elemento a su cluster

```python
labels = kmeans.predict(X)


from mpl_toolkits.mplot3d import Axes3D
#Pintado de elementos y centroides
colores=['red','green','blue','cyan','yellow', 'black']
asignar=[]
for row in labels:
    asignar.append(colores[row])

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=asignar,s=60)
ax.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], marker='*', c=colores, s=1000)
```

## Ejemplo NDimensional

### Entrenamiento
#### Carga de datos

```python
__fichero = "./datasets/datos_clustering.csv"
dataframe = pd.read_csv(__fichero)
print (dataframe.head(5))
dataframe.fillna(0, inplace=True)
X = np.array(dataframe[["H0","H1","H2", "H3","H4","H5", "H6", "H7", "H8", "H9", "H10", "H11", "H12", "H13", "H14", "H15", "H16", "H17", "H18", "H19", "H20", "H21", "H22", "H23"]])
```

#### Estimación del valor de K
Realizamos una preimera pasada hasta un valor de k elevado, con saltos grandes.

```python
__K_min = 1
__K_max = 500
__K_salto=20

Nc = range(__K_min, __K_max, __K_salto)
kmeans = [KMeans(n_clusters=i) for i in Nc]
score = [kmeans[i].fit(X).inertia_ for i in range(len(kmeans))]

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

plt.plot(Nc,score)
plt.xlabel('K')
plt.ylabel('Score')
plt.show()
```

Refinamos la ejecución en el rango [1, 100], con pequeños saltos. Ya que en este rango es donde se consigue el descenso del error

```python
__K_min = 1
__K_max = 100
__K_salto=1

Nc = range(__K_min, __K_max, __K_salto)
kmeans = [KMeans(n_clusters=i) for i in Nc]
score = [kmeans[i].fit(X).inertia_ for i in range(len(kmeans))]

plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

plt.plot(Nc,score)
plt.xlabel('K')
plt.ylabel('Score')
plt.show()
```

### Ejecución
